```python
# -*- coding: utf-8 -*-
"""
Portfolio Construction & Sentiment-Based Stock Analysis
---------------------------------------------------------
This scriptทำการ:
  1. mount Google Drive และโหลดข้อมูลข่าวจาก CSV
  2. ทำความสะอาดข้อมูลข่าวและประมวลผลข้อความด้วย NLTK
  3. วิเคราะห์ Sentiment ด้วย VADER และกำหนด Label
  4. บันทึกข้อมูลที่ผ่านการวิเคราะห์ลงไฟล์ CSV
  5. ดึงข้อมูลราคาหุ้นจาก Yahoo Finance โดยใช้ yfinance
  6. ปรับพอร์ตการลงทุนด้วย PyPortfolioOpt โดยใช้วิธี MVO, HRP, และ MCVaR
  7. จำลองการลงทุนและแสดงผลลัพธ์ (รวมถึงกราฟแสดงการลงทุน, กำไร/ขาดทุน, สัดส่วนการลงทุน และมูลค่าพอร์ตตามเวลา)

@author: Your Name
@version: 1.1
"""

# %% 1. Mount Google Drive & โหลดข้อมูลข่าวสาร

from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import re

# กำหนดประเภทข้อมูลสำหรับบางคอลัมน์
dtypes = {"Title": "string", "Summary": "string", "Ticker": "string"}

# โหลดข้อมูลข่าวจาก CSV โดยเลือกเฉพาะคอลัมน์ที่ต้องการ
csv_path = "/content/drive/MyDrive/Data/filtered_news_sp500_3y.csv"
news_df = pd.read_csv(csv_path, dtype=dtypes, usecols=["Title", "Summary", "Ticker"])

# แสดงข้อมูลดิบ (optional)
print("Initial data shape:", news_df.shape)

# กำจัดแถวที่มีค่าว่างในคอลัมน์ Title หรือ Summary
news_df.dropna(subset=["Title", "Summary"], inplace=True)

# กำจัดแถวที่ซ้ำกันในคอลัมน์ Title และ Summary
news_df.drop_duplicates(subset=["Title", "Summary"], inplace=True)
print("Data shape after cleaning:", news_df.shape)

# %% 2. เตรียมและทำความสะอาดข้อความ (Text Preprocessing)

# Import libraries ที่เกี่ยวข้องกับการประมวลผลภาษา (NLP)
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment import SentimentIntensityAnalyzer

# ตรวจสอบและดาวน์โหลด resource ที่จำเป็นสำหรับ NLTK หากยังไม่มีอยู่
nltk_resources = {
    'stopwords': 'corpora/stopwords',
    'vader_lexicon': 'vader_lexicon',
    'wordnet': 'corpora/wordnet'
}

for resource, path in nltk_resources.items():
    try:
        nltk.data.find(path)
    except LookupError:
        nltk.download(resource)

# กรอกข้อมูล NaN ให้เป็น string ว่าง (เพื่อป้องกัน error ในการประมวลผลข้อความ)
news_df['Title'] = news_df['Title'].fillna('')
news_df['Summary'] = news_df['Summary'].fillna('')

# สร้างคอลัมน์ "Cleaned Text" โดยรวม Title และ Summary เข้าด้วยกัน
news_df['Cleaned Text'] = news_df['Title'] + " " + news_df['Summary']

# กำจัด URL, username (@...) โดยใช้ regular expression
news_df['Cleaned Text'] = news_df['Cleaned Text'].apply(
    lambda x: re.sub(r'http\S+|www\S+|@\S+', '', x)
)

# กำจัดอักขระที่ไม่ใช่ตัวอักษรและเว้นวรรค โดยคงเหลือแค่ตัวอักษร a-z และเว้นวรรค
news_df['Cleaned Text'] = news_df['Cleaned Text'].apply(
    lambda x: re.sub(r'[^a-zA-Z\s]', '', x)
)

# แปลงข้อความให้เป็นตัวพิมพ์เล็ก (lowercase)
news_df['Cleaned Text'] = news_df['Cleaned Text'].apply(lambda x: x.lower())

# กำจัด stopwords (คำที่ไม่สำคัญในการวิเคราะห์)
stop_words = set(stopwords.words('english'))
news_df['No_StopWords'] = news_df['Cleaned Text'].apply(
    lambda x: ' '.join([word for word in x.split() if word not in stop_words])
)

# ทำ Lemmatization เพื่อแปลงคำให้อยู่ในรูปพื้นฐาน โดยใช้ WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
news_df['Lemmatized Text'] = news_df['No_StopWords'].apply(
    lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()])
)

# กำจัดแถวที่อาจจะมี missing values หลังการแปลง (ถ้ามี)
news_df.dropna(subset=['Lemmatized Text'], inplace=True)

# %% 3. วิเคราะห์ Sentiment ด้วย VADER และกำหนด Label

# สร้างตัววิเคราะห์ sentiment
sia = SentimentIntensityAnalyzer()

# คำนวณ sentiment score แบบ compound สำหรับแต่ละข่าว
news_df['Vader_Score'] = news_df['Lemmatized Text'].apply(
    lambda x: sia.polarity_scores(x)['compound']
)

# กำหนด Label ตามค่า sentiment score ด้วยเงื่อนไขที่กำหนดไว้
def set_label(score):
    if score >= 0.5:
        return 'Strong Buy'
    elif score > 0:
        return 'Buy'
    elif score == 0:
        return 'Hold'
    elif score > -0.5:
        return 'Sell'
    else:
        return 'Strong Sell'

news_df['Label'] = news_df['Vader_Score'].apply(set_label)

# บันทึก DataFrame ที่ผ่านการวิเคราะห์ sentiment ลงไฟล์ CSV ใน Google Drive
output_path = '/content/drive/MyDrive/Dat_vader.csv'
news_df.to_csv(output_path, index=False)
print("Sentiment analysis results saved to:", output_path)

# สรุปค่าเฉลี่ย Vader score ต่อ Ticker
avg_sentiment = news_df.groupby('Ticker')['Vader_Score'].mean().reset_index()
avg_sentiment.columns = ['Ticker', 'Average_Vader_Score']
print("Average sentiment score per ticker:")
display(avg_sentiment.sort_values(by='Average_Vader_Score', ascending=False))

# คัดกรองเฉพาะหุ้นที่มีค่าเฉลี่ย sentiment score มากกว่า 0.5 (Strong Buy)
strong_tickers = avg_sentiment[avg_sentiment['Average_Vader_Score'] > 0.5]
print("Tickers with strong sentiment:")
display(strong_tickers)

# %% 4. ดึงข้อมูลหุ้นจาก Yahoo Finance ด้วย yfinance

# ติดตั้ง yfinance หากยังไม่มี
!pip install yfinance --quiet

import yfinance as yf

# แปลง Ticker เป็น list สำหรับดึงข้อมูล
tickers = strong_tickers["Ticker"].tolist()

# กำหนดช่วงเวลาของข้อมูลหุ้น (ตัวอย่าง: 2022-01-01 ถึง 2024-12-31)
start_date = "2022-01-01"
end_date = "2024-12-31"

# ดึงข้อมูลราคาหุ้น โดยใช้ yf.download แล้วเก็บข้อมูลทั้งหมด (OHLC & Volume)
stock_data = yf.download(tickers, start=start_date, end=end_date)
print("Stock data downloaded:")
display(stock_data)

# %%
# ปรับรูปแบบข้อมูลสำหรับการใช้งานต่อไป
# Convert multi-indexed DataFrame to flat DataFrame (reset index)
stock_data_reset = stock_data.stack(level=1).reset_index()
print("Stock data after resetting index:")
display(stock_data_reset)

# %% 5. การปรับพอร์ตการลงทุนด้วย PyPortfolioOpt

# ติดตั้ง PyPortfolioOpt หากยังไม่มี
!pip install PyPortfolioOpt --quiet

import numpy as np
from pypfopt.efficient_frontier import EfficientFrontier, EfficientCVaR
from pypfopt.risk_models import CovarianceShrinkage
from pypfopt.expected_returns import mean_historical_return
from pypfopt.hierarchical_portfolio import HRPOpt
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices

# ใช้ข้อมูลราคา "Close" สำหรับการคำนวณ
# ดึงเฉพาะข้อมูลราคาปิดและลบหุ้นที่ไม่มีข้อมูลทั้งหมด
stock_close = yf.download(tickers, start="2022-01-01", end="2024-12-31")["Close"]
stock_close.dropna(axis=1, how="all", inplace=True)
selected_tickers = stock_close.columns.tolist()  # รายชื่อหุ้นที่มีข้อมูลครบ

# คำนวณผลตอบแทนรายวัน
returns = stock_close.pct_change().dropna()

# คำนวณ expected return (historical return) และ covariance matrix
mu = mean_historical_return(stock_close)
S = CovarianceShrinkage(stock_close).ledoit_wolf()

# --- Portfolio Optimization Models ---

# 1️⃣ MVO (Maximize Sharpe Ratio)
ef_mvo = EfficientFrontier(mu, S, weight_bounds=(0, 1))
mvo_weights = ef_mvo.max_sharpe()
mvo_ret, mvo_risk, mvo_sharpe = ef_mvo.portfolio_performance()

# 2️⃣ HRP (Hierarchical Risk Parity)
hrp = HRPOpt(returns)
hrp_weights = hrp.optimize()
hrp_ret, hrp_risk, hrp_sharpe = hrp.portfolio_performance()

# 3️⃣ MCVaR (Minimize Conditional Value at Risk)
ef_cvar = EfficientCVaR(mu, S, weight_bounds=(0, 1))
cvar_weights = ef_cvar.min_cvar()
cvar_ret, _ = ef_cvar.portfolio_performance()
# คำนวณความผันผวน (volatility) สำหรับ MCVaR
cvar_weights_array = np.array(list(cvar_weights.values()))
cvar_volatility = np.sqrt(np.dot(cvar_weights_array, np.dot(S, cvar_weights_array.T)))
cvar_sharpe = cvar_ret / cvar_volatility

# สรุปผลการเปรียบเทียบพอร์ต
df_comparison = pd.DataFrame({
    "Model": ["MVO (Max Sharpe)", "HRP", "MCVaR"],
    "Expected Return": [mvo_ret, hrp_ret, cvar_ret],
    "Risk (Volatility)": [mvo_risk, hrp_risk, cvar_volatility],
    "Sharpe Ratio": [mvo_sharpe, hrp_sharpe, cvar_sharpe]
})

# แสดงตารางเปรียบเทียบผลการปรับพอร์ต
from IPython.display import display
print("Portfolio Optimization Comparison:")
display(df_comparison.style.format({
    "Expected Return": "{:.2%}",
    "Risk (Volatility)": "{:.2%}",
    "Sharpe Ratio": "{:.2f}"
}))

# บันทึกผลการเปรียบเทียบลง CSV
df_comparison.to_csv("portfolio_comparison.csv", index=False)
print("Portfolio comparison saved as portfolio_comparison.csv")

# %% 6. จำลองการลงทุนโดยใช้พอร์ต MVO (Max Sharpe)

# กำหนดเงินลงทุนเริ่มต้น
initial_capital = 10000  # USD

# กำหนดช่วงเวลาสำหรับจำลองการลงทุน (ตัวอย่าง: มกราคม 2025)
simulation_start = "2025-01-01"
simulation_end = "2025-01-31"

# ดึงข้อมูลราคาหุ้นในช่วงเวลาจำลองโดยใช้พอร์ตที่เลือกจาก MVO
tickers_mvo = list(mvo_weights.keys())
simulation_data = yf.download(tickers_mvo, start=simulation_start, end=simulation_end)['Close']

# เติมค่า NaN ด้วยการ forward fill เพื่อความต่อเนื่องของข้อมูล
simulation_data = simulation_data.fillna(method="ffill")

# กำหนดราคาเริ่มต้นและราคาสิ้นสุดของหุ้นแต่ละตัว
start_prices = simulation_data.iloc[0]
end_prices = simulation_data.iloc[-1]

# กรองหุ้นที่มีข้อมูลในวันแรก
valid_tickers = start_prices.dropna().index.tolist()
start_prices = start_prices[valid_tickers]
end_prices = end_prices[valid_tickers]

# คำนวณจำนวนเงินลงทุนในแต่ละหุ้น (ตามสัดส่วนที่ได้จาก MVO)
investment_per_stock = {ticker: initial_capital * mvo_weights[ticker] for ticker in valid_tickers}

# คำนวณจำนวนหุ้นที่ซื้อได้ โดยใช้ราคาขึ้นต้น
shares_bought = {ticker: investment_per_stock[ticker] / start_prices[ticker] for ticker in valid_tickers}

# คำนวณมูลค่าพอร์ตในวันสิ้นสุดการจำลอง
final_values = {ticker: shares_bought[ticker] * end_prices[ticker] for ticker in valid_tickers}
final_portfolio_value = sum(final_values.values())

# คำนวณกำไร/ขาดทุนและ ROI
profit_loss = final_portfolio_value - initial_capital
roi = (profit_loss / initial_capital) * 100

print("=== Investment Simulation Results ===")
print(f"Initial Investment: ${initial_capital:,.2f}")
print(f"Final Portfolio Value: ${final_portfolio_value:,.2f}")
print(f"Profit/Loss: ${profit_loss:,.2f} ({roi:.2f}%)")

# %% 7. แสดงผลการจำลองการลงทุนด้วยกราฟ

# สร้าง DataFrame สรุปข้อมูลการลงทุนในแต่ละหุ้น
summary_df = pd.DataFrame({
    "Stock": valid_tickers,
    "Shares Bought": [shares_bought[ticker] for ticker in valid_tickers],
    "Start Price": [start_prices[ticker] for ticker in valid_tickers],
    "End Price": [end_prices[ticker] for ticker in valid_tickers],
    "Investment": [investment_per_stock[ticker] for ticker in valid_tickers],
    "Final Value": [final_values[ticker] for ticker in valid_tickers]
})

# กรองเฉพาะหุ้นที่มีการลงทุน (Investment > 0)
summary_df = summary_df[summary_df["Investment"] > 0]
summary_df["Profit/Loss"] = summary_df["Final Value"] - summary_df["Investment"]

print("=== Summary of Stock Investments ===")
display(summary_df)

import matplotlib.pyplot as plt

# --- Bar Chart: Investment vs. Profit/Loss per Stock ---
fig, ax = plt.subplots(figsize=(12, 6))

# แสดงกราฟแท่งสำหรับการลงทุน (Investment)
bars1 = ax.bar(summary_df["Stock"], summary_df["Investment"],
               label="Investment", color="royalblue", alpha=0.7)

# แสดงกราฟแท่งสำหรับกำไร/ขาดทุน (Profit/Loss)
bars2 = ax.bar(summary_df["Stock"], summary_df["Profit/Loss"],
               label="Profit/Loss",
               color=summary_df["Profit/Loss"].apply(lambda x: "green" if x > 0 else "red"))

# แสดงตัวเลขบนแต่ละแท่ง
for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, f"${yval:,.2f}",
            ha='center', va='bottom', fontsize=10, color="black")
for bar in bars2:
    yval = bar.get_height()
    # วางตำแหน่งข้อความขึ้นอยู่กับการเป็นกำไรหรือขาดทุน
    if yval >= 0:
        ax.text(bar.get_x() + bar.get_width()/2, yval, f"+${yval:,.2f}",
                ha='center', va='bottom', fontsize=10, color="black")
    else:
        ax.text(bar.get_x() + bar.get_width()/2, yval, f"${yval:,.2f}",
                ha='center', va='top', fontsize=10, color="black")

ax.set_xlabel("Stocks", fontsize=12)
ax.set_ylabel("Value (USD)", fontsize=12)
ax.set_title("Investment and Profit/Loss for Each Stock", fontsize=14)
ax.legend()
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# --- Pie Chart: สัดส่วนการลงทุนในแต่ละหุ้น ---
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(summary_df["Investment"], labels=summary_df["Stock"], autopct="%1.1f%%", startangle=90)
ax.set_title("Investment Allocation by Stock")
plt.show()

# --- Line Chart: การเปลี่ยนแปลงมูลค่าพอร์ตโฟลิโอตามเวลา ---
# คำนวณมูลค่าพอร์ตโฟลิโอเป็นช่วงเวลาจำลองโดยคูณหุ้นที่ซื้อกับราคาปิดในแต่ละวัน
portfolio_values = simulation_data[summary_df["Stock"]].values @ summary_df["Shares Bought"].values

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(simulation_data.index, portfolio_values, label="Portfolio Value", color="blue", linewidth=2)
ax.set_xlabel("Date")
ax.set_ylabel("Portfolio Value (USD)")
ax.set_title("Portfolio Value Over Time")
ax.legend()
ax.grid(True)
plt.show()
```
